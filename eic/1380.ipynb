```python
%sql
select 
  date_ym
  , count(distinct mac_addr) as device_cnt
  , count(distinct X_User_Number) as user_cnt
  , count(distinct concat(mac_addr, X_User_Number)) as device_cnt
  , count(1) as row_cnt 
from eic_data_ods.tlamp.normal_log_webos25
where 1=1
  and date_ym between '2025-04' and '2026-04'
  and context_name ='BROADCAST'
  and message_id = 'NL_BROADCAST_WATCH_CHANNELSOURCE'
  and normal_log:atv>=1
group by 
  date_ym
order by
  date_ym
```
