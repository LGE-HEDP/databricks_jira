```sql

SELECT
    X_device_platform as `platform_code`,  -- 필수 열
    X_device_product as `platform_version`,-- 필수 열
    X_Device_Country as `country_code`,    -- 필수/조건 열
    log_create_time,                       -- 필수/조건 열
    X_Device_SDK_VERSION as `dpv`,
    X_Device_FW_Version,
    -- mac_addr as `device_id`,               -- 필수 열  
    x_device_sales_model as `sales_model_code`, -- 필수 열
    context_name,                          -- 필수/조건 열  
    message_id,                            -- 필수/조건 열
    normal_log:top10
FROM eic_data_ods.tlamp.normal_log_webos22
WHERE 1=1
    AND X_device_country = 'TR'
    AND context_name = 'MM'
    AND message_id = 'NL_MM_PROCLIST'
    AND date_ym >= '2026-01'
    AND date_ym < '2026-03'
    AND mac_addr = sha2(concat(upper('74:C1:7E:01:99:63'), secret('admin', 'salt')), 256)
;
```
