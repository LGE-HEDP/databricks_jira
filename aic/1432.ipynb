```python
%sql
select
    webos_version,
    `0`,
    `1`,
    `2`,
    `null`
from (
    select
        case
            when X_Device_Platform like 'W21%' then 'webOS60'
            when X_Device_Platform like 'W26%' then 'webOS26'
            when X_Device_Platform like 'W25%' then 'webOS25'
            when X_Device_Platform like 'W24%' then 'webOS24'
            when X_Device_Platform like 'W23%' then 'webOS23'
            when X_Device_Platform like 'W22%' then 'webOS22'
            else 'etc'
        end as webos_version,
        coalesce(cast(normal_log:vrr_enabled as string), 'null') as vrr_enabled_value,
        count(distinct mac_addr) as mac_count
    from (
        select * from aic_data_ods.tlamp.normal_log_webos60
        union all
        select * from aic_data_ods.tlamp.normal_log_webos26
        union all
        select * from aic_data_ods.tlamp.normal_log_webos25
        union all
        select * from aic_data_ods.tlamp.normal_log_webos24
        union all
        select * from aic_data_ods.tlamp.normal_log_webos23
        union all
        select * from aic_data_ods.tlamp.normal_log_webos22
    ) t
    where 1=1
        -- and date_ym between '2026-06' and '2026-06'
        and context_name = 'com.webos.service.extinputs.hdmi'
        and message_id = 'NL_EXTINPUTS_HDMI_STATUS'
        and PRODUCT_NUM like 'OLED48%'
        and (
            X_Device_Platform like 'W21%' or
            X_Device_Platform like 'W26%' or
            X_Device_Platform like 'W25%' or
            X_Device_Platform like 'W24%' or
            X_Device_Platform like 'W23%' or
            X_Device_Platform like 'W22%'
        )
    group by webos_version, normal_log:vrr_enabled
) src
pivot (
    sum(mac_count) for vrr_enabled_value in ('0', '1', '2', 'null')
)
order by webos_version
```
