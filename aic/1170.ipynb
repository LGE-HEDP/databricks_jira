```python
%sql
with raw_data as (
  SELECT 
      X_device_platform as `platform_code`,  -- 필수 열
      X_device_product as `platform_version`,-- 필수 열
      X_Device_Country as `country_code`,    -- 필수/조건 열
      log_create_time,                       -- 필수/조건 열
      X_Device_SDK_VERSION as `dpv`,
      mac_addr,               -- 필수 열  
      x_device_sales_model as `sales_model_code`, -- 필수 열
      context_name,                          -- 필수/조건 열  
      message_id,                            -- 필수/조건 열
      normal_log:activate_type, -- {"activate_type":"concierge","type":"recommend","keyword":"A Minecraft Movie"}
      normal_log:type as log_type,
      normal_log:keyword
  FROM kic_data_ods.tlamp.normal_log_webos25
  WHERE 1=1
      AND X_device_country = 'KR'
      and context_name = 'aiPlatform'
      and message_id = 'NL_SELECT_CONCIERGE_KEYWORD' -- 'NL_SELECT_CONTENT'
      AND X_device_product = 'webOSTV 25'
      AND (x_device_sales_model LIKE '65QNED85AKS%' OR x_device_sales_model LIKE '75QNED85AKS%') -- EASY TV 모델 --  : 65QNED85AKS/75QNED85AKS)
      AND date_ym >= '2025-10'
      AND date_ym <= '2025-12'  -- `25.10.01 ~ `25.11.09
), fact_total_ud_by_job_and_app as (
  select activate_type, log_type, count(distinct mac_addr) as total_ud
  from   raw_data 
  group by activate_type, log_type 
), fact_count_ud_by_job_and_app as (
  select activate_type, keyword, log_type, count(mac_addr) as count_log, count(distinct mac_addr) as count_ud
  from   raw_data
  group by activate_type, keyword, log_type
  order by activate_type, keyword, log_type
), fact_total_ud_by_app as (
  select count(distinct mac_addr) as total_ud
  from   raw_data
), fact_count_ud_by_app as (
  select 'Total' as activate_type, keyword, log_type, count(mac_addr) as count_log, count(distinct mac_addr) as count_ud
  from   raw_data 
  group by keyword, log_type 
)

select * 
from (
  (select count.*, total.total_ud
  from   fact_count_ud_by_job_and_app count
  join   fact_total_ud_by_job_and_app total 
  on count.activate_type = total.activate_type and count.log_type = total.log_type
  order by count.activate_type, count.keyword)

  union all 

  (select * from fact_count_ud_by_app, fact_total_ud_by_app)
) t
```
