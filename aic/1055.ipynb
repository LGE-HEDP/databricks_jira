```python
import io
import boto3
import pandas as pd

s3 = boto3.client('s3',
                  region_name='ap-northeast-2',
                  aws_access_key_id='',
                  aws_secret_access_key='')

bucket_name = 's3-lge-he-inbound-kic-dev'
file_path = 'HEDS/HEDS-1055/IDC_20252Q.csv'

obj = s3.get_object(Bucket=bucket_name, Key=file_path)
df = pd.read_csv(io.BytesIO(obj['Body'].read()))
```

```python
print(df)
```

```python
%python
# Rename columns to replace spaces and special characters with underscores
df_renamed = df.rename(
    columns=lambda x: x.strip()
    .replace(" ", "_")
    .replace(",", "_")
    .replace(";", "_")
    .replace("{", "_")
    .replace("}", "_")
    .replace("(", "_")
    .replace(")", "_")
    .replace("\n", "_")
    .replace("\t", "_")
    .replace("=", "_")
)

spark.createDataFrame(df_renamed).write.mode("overwrite").saveAsTable(
    "sandbox.z_suhyang_kim.IDC_20252Q"
)
```
