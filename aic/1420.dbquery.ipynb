```sql
SELECT DISTINCT
  COALESCE(dim.`app_name_en`, sdp.`TITLE`) AS app_name,
  sdp.`STR_APP_ID` AS app_id,
  CASE sdp.`CAT_CODE`
    WHEN 2 THEN 'Game'
    WHEN 3 THEN 'Entertainment'
    WHEN 4 THEN 'Life'
    WHEN 5 THEN 'Education'
    WHEN 6 THEN 'Information'
    ELSE 'Other'
  END AS category_name
FROM `aic_data_ods`.`tlamp`.`sdp_app_master` sdp
INNER JOIN `aic_data_ods`.`tlamp`.`app_install_history` hist
  ON sdp.`STR_APP_ID` = hist.`app_id`
LEFT JOIN `aic_data_dimension`.`common_tv`.`app_id_code` dim
  ON sdp.`STR_APP_ID` = dim.`app_id`
WHERE (hist.`Country_code` = 'US'
  OR hist.`Country_code` = 'CA')
  AND sdp.`USE_FLAG` = 'Y'
  AND sdp.`CAT_CODE` IS NOT NULL
ORDER BY category_name, app_name
;
```
