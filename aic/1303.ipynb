```python
%sql
select date_ym, platform_version, count(1)
from aic_data_mart.tv_usage.hdmi_usage
where 1=1
  and date_ym between'2025-01' and '2026-01'
  and platform_version in ('webOS6.0','webOS24')
  and (
    sales_model_code like '%27ART10%'
    OR sales_model_code like '%27LX6T%'
    OR (sales_model_code like '%C5%' and sales_model_code like '%OLED%')
    OR sales_model_code like '%QNED85%'
  )
  GROUP BY date_ym, platform_version
  ORDER BY 1,2
```

```python
%sql
select * except (rn)
from (
  select *,
         row_number() over (order by date_ym) as rn
  from aic_data_mart.tv_usage.hdmi_usage
  where 1=1
    and date_ym between '2025-01' and '2026-02'
    and platform_version in ('webOS6.0','webOS24')
    and (
      sales_model_code like '%27ART10%'
      OR sales_model_code like '%27LX6T%'
      OR (sales_model_code like '%C5%' and sales_model_code like '%OLED%')
      OR sales_model_code like '%QNED85%'
    )
) t
where rn between 1 and 800000
order by rn
```

```python
%sql
select * except (rn)
from (
  select *,
         row_number() over (order by date_ym) as rn
  from aic_data_mart.tv_usage.hdmi_usage
  where 1=1
    and date_ym between '2025-01' and '2026-02'
    and platform_version in ('webOS6.0','webOS24')
    and (
      sales_model_code like '%27ART10%'
      OR sales_model_code like '%27LX6T%'
      OR (sales_model_code like '%C5%' and sales_model_code like '%OLED%')
      OR sales_model_code like '%QNED85%'
    )
) t
where rn between 800001 and 1600000
order by rn
```
