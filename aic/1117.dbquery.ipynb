```sql
select 
  substr(log_create_time,1,10) as date_ymd,
  context_name,
  message_id,
  cntry.region_name as region_name,
  cntry.country_name as country_name,
  normal_log.mac_addr as device_id,
  normal_log.X_Device_Product as platform_version,
  normal_log.normal_log:category_id as category_id,
  normal_log.normal_log:sub_id as sub_id,
  normal_log.normal_log:use_time as use_time,
  count(1) as access_count
from kic_data_ods.tlamp.normal_log_webos22 normal_log
  inner join kic_data_dimension.common.country_code cntry
  on normal_log.X_Device_Country = cntry.country_code
where 1=1
-- and date_ym = '2024-07'
and log_create_time >= '2024-07-01'
and log_create_time < '2024-08-01'
and X_Device_Country = 'KR'
and context_name = 'com.webos.app.lifeonscreen'
and message_id = 'NL_CATEGORY_USE_TIME'
group by all
```
