```python
%sql
select 
  X_Device_Product, context_name, message_id, count(1) row_cnt, count(distinct mac_addr) as mac_cnt
from kic_data_ods.tlamp.normal_log_webos25
where 1=1
  and date_ym = '2025-12'
  and X_Device_Country = 'IN'
  and (
    (context_name = 'com.webos.app.homeconnect' and message_id = 'NL_HC_INPUTLIST')
    or (context_name = 'EIM' and message_id = 'NL_EIM_INPUTLIST')
    or (context_name = 'ICS' and message_id = 'NL_UVC_LIST')
    or (context_name = 'com.webos.app.home' and message_id in ('NL_SHELF_LIST_SHOWN', 'NL_HOME_SHOWN', 'NL_HERO_SHOWN', 'NL_APP_LAUNCH'))
    or (context_name = 'com.webos.app.livedmost' and message_id = 'NL_LGCHANNELS')
  )
  and X_Device_Product = 'webOSTV 25'
group by 
  X_Device_Product, context_name, message_id
order by 
  1,2,3
```
